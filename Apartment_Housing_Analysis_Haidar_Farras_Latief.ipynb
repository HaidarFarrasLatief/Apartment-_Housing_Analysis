## 2.5 Glossary

### 2.5.1 Glossary Istilah diluar kolom dataset



```
1. BHK:
    merupakan singkatan dari "Bedroom, Hallroom, Kitchen". Angka-angka yang tertera pada kolom BHK ini mengandung arti: jika semakin tinggi angkanya, maka semakin banyak "Bedroom"/ruang tidurnya, namun tidak seiring naik jumlah pada "Hallroom"/ruang tamu dan "Kitchen"/dapur.

    1 BHK artinya: 1 Bedroom, 1 Hallroom, 1 Kitchen

    Naiknya jumlah BHK hanya menaikan jumlah Bedroomnya saja, Hallroom dan Kitchen tetap pada jumlah 1.

    BHK bisa mempunyai jumlah yang desimal, itu mengartikan ada 1 Bedroom yang mempunyai ukurang yang kurang dari bedroom lainnya. Namun, pada dataset ini, tidak ditemukan BHK yang mempunyai nilai desimal. 

    Untuk informasi lebih lanjut mengenai BHK, bisa ditelusuri melewati link ini: 
    https://www.magicbricks.com/blog/what-is-bhk-full-form/126037.html#:~:text=BHK%20stands%20for%20'Bedroom%2C%20Hall,room)%2C%20and%20one%20Kitchen.
```

### 4.2.6 Penjabaran 6
```
Apakah ada hubungan Area Type dengan Point of Contact ?
```

Untuk melakukan test ini, saya gunakan hipotesis:
```
H0: Area Type dan Point of Contact tidak mempunyai hubungan
H1: Area Type dan Point of Contact mempunyai hubungan
```
Dikarenakan kedua variabel ini mempunyai isi konten string atau merupakan data ketegori, maka saya gunakan Chi Square Test

Untuk melakukan Chi Square Test, diperlukan 4 asumsi:

1. Kedua variabel adalah kategori

2. Semua observasi tidak bergantung (Independent)

3. Sel atau isi konten dari tabel kontigensi sama-sama eksklusif atau isi konten yang melakukan pengulangan

4. Expected value pada masing-masing sel harus berjumah 5 atau lebih sekiranya 80% dari cel

````python
contingency_table_contoh = pd.crosstab(df_copy["Area Type"], df_copy["Point of Contact"])
contingency_table_contoh
````

Pada tabel diatas ini terlihat bahwa terdapat sel yang berada pada bawah 5. Walaupun saya hitung expected valuenya, ada sel yang mendapatkan hasil dibawah 5, maka itu saya keluarkan kolom 'Contact Builder' dan baris 'Built Area' sehingga tertinggal angka yang bisa diolah pada Chi Square test ini.

````python
df_chi_1 = df_copy[df_copy["Point of Contact"].str.contains("Contact Builder") == False]
df_chi_2 = df_chi_1[df_chi_1["Area Type"].str.contains("Built Area") == False]

contingency_table = pd.crosstab(df_chi_2["Area Type"], df_chi_2["Point of Contact"])
contingency_table 
````

Dengan filter data ini, data dapat dimasukan kedalam Chi Square Test namun dengan informasi tambahan bahwa ada data yang berkurang.

````python
# Melakukan test chi
stat, p, dof, expected = stats.chi2_contingency(contingency_table)
print('p=%.12f' % (p))
if p > 0.05:
  print("Probably Independent")
else: 
  print("Probably Dependent")
````

Hasil menunjukan untuk menolak H0 yang berarti: Area Type dan Point of Contact mempunyai hubungan.
